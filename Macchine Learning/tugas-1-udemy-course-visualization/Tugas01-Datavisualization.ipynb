```python
import pandas as pd
# import csv and look if its working fine 
courses = pd.read_csv("udemy_courses.csv")
courses.head(5)
```

```python
# Drop non Web Development courses
indexNames = courses[(courses.subject != "Web Development")].index
courses.drop(indexNames,inplace=True)

# Drop free courses
indexNames = courses[(courses.is_paid == False)].index
courses.drop(indexNames,inplace=True)

courses.head(5)
```

```python
print("Price mean       : ", courses.price.mean())
print("Subscribers mean : ", courses.num_subscribers.mean())
print("Duration mean    : ", courses.content_duration.mean())
```

```python
print("Price median       : ", courses.price.median())
print("Subscribers median : ", courses.num_subscribers.median())
print("Duration median    : ", courses.content_duration.median())
```

```python
print("\nPrice modus       :\n", courses.price.mode())
print("\nSubscribers modus :\n", courses.num_subscribers.mode())
print("\nDuration modus    :\n", courses.content_duration.mode())
```

```python
print("Price varian       : ", courses.price.var())
print("Subscribers varian : ", courses.num_subscribers.var())
print("Duration varian    : ", courses.content_duration.var())
```

```python
print("Price standar deviasi       : ", courses.price.std())
print("Subscribers standar deviasi : ", courses.num_subscribers.std())
print("Duration standar deviasi    : ", courses.content_duration.std())
```

```python
# Show some feature graphs
courses.hist(column=['price','num_subscribers','content_duration'],facecolor='Blue',figsize=(25,24),layout=(4,2),alpha=0.65,bins=300);   
```

```python
print("Correlation between course duration and subscribers   : ", courses.content_duration.corr(courses.num_subscribers))
print("Correlation between course price and subscribers      : ", courses.price.corr(courses.num_subscribers))
print("Correlation between course reviews and subscribers    : ", courses.num_reviews.corr(courses.num_subscribers))
```
